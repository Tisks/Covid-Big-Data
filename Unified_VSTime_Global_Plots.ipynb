```
import pandas as pd
import matplotlib.pyplot as plt

def readAndPlot(data, x_label, y_label, title, legend, color):
  # data manipulation
  data = data.drop(['Province/State','Country/Region','Lat','Long'], axis=1)
  data = data[data.columns].sum()

  # plot
  f = plt.figure()
  plt.ticklabel_format(style = 'plain')
  plt.title(title)
  plt.xlabel(xlabel=x_label)
  plt.ylabel(ylabel=y_label)
  plt.grid(True)
  
  dataPlot = data.plot(figsize=(10,5), color=color, ax=f.gca())
  dataPlot.legend([legend], loc='upper left')
  dataPlot.tick_params(axis='x', labelrotation=90)

  plt.show()
```

```
# data download

## contagiados vs tiempo
dfHum1 = pd.read_csv('https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_confirmed_global.csv&filename=time_series_covid19_confirmed_global.csv', sep=',')
dfHopkins1 = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv', sep=',')

## recuperados vs tiempo
dfHum2 = pd.read_csv('https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_recovered_global.csv&filename=time_series_covid19_recovered_global.csv', sep=',')
dfHopkins2 = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv', sep=',')


## fallecidos vs tiempo
dfHum3 = pd.read_csv('https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_deaths_global.csv&filename=time_series_covid19_deaths_global.csv', sep=',')
dfHopkins3 = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv', sep=',')
```

```
# plot Contagiados v/s tiempo: HumData

readAndPlot(data=dfHum1, x_label='Tiempo', y_label='Contagiados', title='Contagiados v/s tiempo: HumData', legend='Contagiados por día', color='orange')
```

```
# plot Contagiados v/s tiempo: Johns Hopkins University CSSE

readAndPlot(data=dfHopkins1, x_label='Tiempo', y_label='Contagiados', title='Contagiados v/s tiempo: Johns Hopkins University CSSE', legend='Contagiados por día', color='y')
```

```
# plot Recuperados v/s tiempo: HumData

readAndPlot(data=dfHum2, x_label='Tiempo', y_label='Recuperados', title='Recuperados v/s tiempo: HumData', legend='Recuperados por día', color='g')
```

```
# plot Recuperados v/s tiempo: Johns Hopkins University CSSE

readAndPlot(data=dfHopkins2, x_label='Tiempo', y_label='Recuperados', title='Recuperados v/s tiempo: Johns Hopkins University CSSE', legend='Recuperados por día', color='b')
```

```
# plot Fallecidos v/s tiempo: HumData

readAndPlot(data=dfHum3, x_label='Tiempo', y_label='Fallecidos', title='Fallecidos v/s tiempo: HumData', legend='Fallecidos por día', color='r')
```

```
# plot Fallecidos v/s tiempo: Johns Hopkins University CSSE

readAndPlot(data=dfHopkins3, x_label='Tiempo', y_label='Fallecidos', title='Fallecidos v/s tiempo: Johns Hopkins University CSSE', legend='Fallecidos por día', color='m')
```
