```
import pandas
import matplotlib.pyplot as plt

recuperadosHum = pandas.read_csv('https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_recovered_global.csv&filename=time_series_covid19_recovered_global.csv', sep=',')
recuperadosJH = pandas.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')

humData = recuperadosHum.drop(['Province/State','Country/Region','Lat','Long'], axis=1)
johnHData = recuperadosJH.drop(['Province/State','Country/Region','Lat','Long'], axis=1)

humData = humData[humData.columns].sum()
# humData

johnHData = johnHData[johnHData.columns].sum()
# johnHData
```

```
import matplotlib.pyplot as plt
```

```
humPlot = humData.plot(figsize=(20,10), color='green')
humPlot.set_title('Recuperados v/s tiempo: HumData')
humPlot.set_xlabel('Tiempo')
humPlot.set_ylabel('Recuperados')
humPlot.tick_params(axis='x', labelrotation=90)
humPlot.legend(['Recuperados por día']);

```

```
johnHPlot = johnHData.plot(figsize=(20,10))
johnHPlot.set_title('Recuperados v/s tiempo: Johns Hopkins University CSSE')
johnHPlot.set_xlabel('Tiempo')
johnHPlot.set_ylabel('Recuperados')
johnHPlot.tick_params(axis='x', labelrotation=90)
johnHPlot.legend(['Recuperados por día']);
```
