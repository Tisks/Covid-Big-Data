```
import pandas
# import matplotlib

df1 = pandas.read_csv('https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_recovered_global.csv&filename=time_series_covid19_recovered_global.csv', sep=',')
df2 = pandas.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')


```

```
humData = df1.drop(['Province/State','Country/Region','Lat','Long'], axis=1)
humData
```

```
johnHData = df2.drop(['Province/State','Country/Region','Lat','Long'], axis=1)
johnHData
```

```
humData = humData[humData.columns].sum()
```

```
johnHData[johnHData.columns].sum()
```

```
humPlot = humData.plot(figsize=(20,5), color='green')
humPlot.set_title('Recuperados v/s tiempo')
humPlot.set_xlabel('Tiempo')
humPlot.set_ylabel('Recuperados')
humPlot.legend(['Recuperados por dÃ­a']);
```
