```
import pandas as pd
import matplotlib.pyplot as plt
```

```
total_nacional = pd.read_csv('https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto5/TotalesNacionales.csv', sep= ',')
total_nacional
```

```
a = 4
b=10
a_title = 'Casos activos'
b_title = 'Casos recuperados'

activos = total_nacional[a:a+1].drop(['Fecha'], axis=1) 
recuperados = total_nacional[b:b+1].drop(['Fecha'], axis=1) 
frames = [activos, recuperados]
result = pd.concat(frames)
result = result.T
result.columns = [a_title ,b_title]




result
```

```


p = result.plot(figsize =(20,10))
```
