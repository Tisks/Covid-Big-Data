```
import pandas as pd
import matplotlib.pyplot as plt
```

```
total_nacional = pd.read_csv('https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto5/TotalesNacionales.csv', sep= ',')
total_nacional
```

```
a = 4

b=10
c=3
d=1
a_title = 'Casos activos'
b_title = 'Casos recuperados'
c_title = 'Fallecidos'
d_title= 'Casos Totales'
fallecidos = total_nacional[c:c+1].drop(['Fecha'], axis=1) 
activos = total_nacional[a:a+1].drop(['Fecha'], axis=1) 
recuperados = total_nacional[b:b+1].drop(['Fecha'], axis=1) 
total= total_nacional[d:d+1].drop(['Fecha'], axis=1) 

frames = [activos, recuperados, fallecidos, total]
#frames = [activos, recuperados,fallecidos]
result = pd.concat(frames)
result = result.T
#result = activos.T
#result.columns = [a_title ,b_title,c_title]

result.columns = [a_title,b_title, c_title, d_title]


result
```

```


p = result.plot(figsize =(20,10))
plt.grid()
plt.title('Active, recovered, total cases and deaths v/s time ',fontsize=20)
plt.ylabel('People',fontsize=15)
plt.xlabel('Date',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```
